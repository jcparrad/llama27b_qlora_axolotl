realizar la evaluacion del set de test. Esta evaluacion es de clasificacion multi label

```python
import pandas as pd
import ast
```

```python
# Function to parse the input string in the 'text_column' and return the parsed dictionary
def parse_input_string(input_string):
    try:
        return ast.literal_eval(input_string)
    except (SyntaxError, ValueError) as e:
        print(f"Error parsing input string: {e}")
        return None
```

```python
df = pd.read_excel("response/response.xlsx")
df['parsed_text'] = df['response_json'].apply(parse_input_string)
df
```

```python
row = df.iloc[0]
row["output"]
```

```python
import re

def extract_elements(input_string):
    # Use regular expression to find the content inside the square brackets
    match = re.search(r'\[([^]]*)\]', input_string)

    if match:
        # Extract and split the content inside the square brackets into a list
        result_list = [item.strip("' ") for item in match.group(1).split()]
        return result_list
    else:
        return None

# Example usage:
result_list = extract_elements(row["output"])
result_list
```

```python
# obteniendo los tags ground-truth multilabel

df["gt_tags_multilabel"] = df.apply(lambda row: extract_elements(row["output"]) ,axis = 1)
df
```

```python
row = df.iloc[0]
row["parsed_text"]["text"][0]
```

```python
# extracting the tags from response


df["pred_tags_multilabel"] = df.apply(lambda row: extract_elements(row["parsed_text"]["text"][0]) ,axis = 1)
df


```

```python
df = df.dropna(subset=['gt_tags_multilabel', 'pred_tags_multilabel'])
df
```

```python
row = df.iloc[0]
row["gt_tags_multilabel"]
```

```python
    import pandas as pd
    from sklearn.metrics import classification_report
    from sklearn.preprocessing import MultiLabelBinarizer


    # Convert the lists in the DataFrame to binary label indicators
    mlb = MultiLabelBinarizer()
    y_true = mlb.fit_transform(df['gt_tags_multilabel'])
    y_pred = mlb.transform(df['pred_tags_multilabel'])

    # Compute classification report
    report = classification_report(y_true, y_pred, target_names=mlb.classes_, output_dict=True)

    # Display the classification report
    print(pd.DataFrame(report).transpose())
```
